```python
import requests
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Odisha').text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(website_url, 'lxml')
#print(soup.prettify())
```

```python
table = soup.find('table', {'class': 'wikitable plainrowheaders sortable mw-collapsible citiwise-data'})
```

```python
table
```

```python
districts = []

links = table.find_all('a')
for link in links:
    #print(link.get('title'))
    districts.append(link.get('title'))

districts = districts[4:33]
```

```python
for i in range(5):
    print(table.find_all('td')[i].contents[0])
```

```python
active = table.find_all('td')[0].contents[0]
print('Active : ', str(active))
```

```python
n = len(districts) * 5
covid_list = []

for i in range(n):
    covid_list.append(table.find_all('td')[i].contents[0])

type(covid_list)
```

```python
data = []
a = 0
b = 5
while a < 145:
    data.append(covid_list[a:b])
    a += 5
    b += 5
    
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=['Active', 'Recovered', 'Deaths', 'First Case', 'Notes'])
```

```python
# Using DataFrame.insert() to add a column 
df.insert(0, "Districts", districts, True) 
```
